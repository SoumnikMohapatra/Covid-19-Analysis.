```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_csv('4. covid_19_data.csv')
data.head()

```

```python
'''Q-1) Show the number of confirmed,death and number of recovered cases.'''
data.groupby('Region').sum()

```

```python
#Q-2) Remove all the records where confirmed cases is less than 20.
data = data[~(data.Confirmed < 20)]
data
```

```python
data.groupby('Region')['Confirmed'].sum().sort_values(ascending=False).head(20) #Q-3) in which region there is maximum no of confirmed cases.
         #Seeing the result we conclude that US has maximum confirmed cases.
```

```python
#Q-4)find the region in which minimum no of deaths reported.
data.groupby('Region')['Deaths'].sum().sort_values(ascending=True).head(20)
```

```python
#Q-5)how many confirmed,recovered & death cases reported in US till 29 April 2020.
data[data.Region == 'US']
```

```python
#Q-6) sort the entire data of confirmed cases in ascending order.
data.sort_values(by='Confirmed',ascending=True)
```

```python
#Q-7)sort the data wrt no of death cases in descending order.
data.sort_values(by='Deaths',ascending=False)
```

```python
import matplotlib.pyplot as plt #Q-8) plot a bar graph of confirmed cases in US,Spain,Italy,France.
plt.bar(['US','Spain','Italy','France'],[1039895,236899,203591,166518],label='Bar',color='r')
plt.xlabel('Region')
plt.ylabel('Confirmed cases')
plt.title('Region vs Confirmed Cases')
plt.show()
```

```python
Region_list = data['Region'].tolist()  #Q-9)Show bar ploth for death cases vs region for all regions.
Death_list = data['Deaths'].tolist()
plt.bar(Region_list,Death_list,label='Bar',color='b')
plt.xlabel('Region')
plt.ylabel('Death cases')
plt.title('Region vs Death cases')
plt.show()
```
